# Maximum-Likelihood Estimation

Run the command below to upload 100 data points $x_1,x_2,...,x_{100}$ concerning individuals' annual salaries.

The sample mean is:
$$\bar{X}=\frac{1}{100}\,\sum_{i=1}^{100} x_i.$$
The sample standard deviation is:
$$s=\sqrt{\frac{1}{99}\,\sum_{i=1}^{100} (x_i-\bar{x})^2}.$$



```r

#Import the CSV file with the data
#The data are collected matrix X with two columns and 100 rows. The sata are collected in the second column, and thus we analyse X[,2]
X<-read.csv("https://www.enricodegiorgi.com/teaching/MatheB/EN/salaries.csv")

#Compute average salary
xbar<-(1/100)*sum(X[,2])
s<-sqrt((1/99)*sum((X[,2]-xbar)^2))

#Prompt xbar and s
print(paste("sample mean=",xbar))
print(paste("sample standard deviation=",s))


#Plot an histogram of the data
df<-hist(X[,2],breaks=100,main="Distribution of salaries",freq=FALSE,xlab="Salaries (CHF)")
```

We assume that data are generated by $n=100$ independent random variables $X_1,\dots,X_n$ that are distributed according to the exponential distribution with parameter $\lambda>0$.

The Likelihood function of the data corresponds to:
$$L(\lambda;x_1,\dots,x_{100})=\prod_{i=1}^{100} \lambda\,e^{-\lambda\,x_i}=\lambda^{100}\,e^{-\lambda\,\sum_{i=1}^{100} x_i}.$$

We apply the natural logarithm to $L(\lambda;x_1,\dots,x_n)$ and obtain:
$$\ln L(\lambda;x_1,\dots,x_{100})=100\,\ln(\lambda)-\lambda\,\sum_{i=1}^{100} x_i.$$

We maximise this function and obtain the Maximum-Liklelihood Estimation of $\lambda$:
$$\hat{\lambda}=\frac{1}{\bar{x}}.$$

Run the code below to display the Log-Likelihod function and its maximum at $\frac{1}{\bar{x}}$.

```r
##Compute the MLE of lambda
lambdahat<-1/xbar

#Prompt lambdahat
print(paste("lambda hat=",lambdahat))

#Plot the Log-likelihood function and its maximum at lambdahat
lambda<-seq(0,0.0001,0.000001)
lnL<-100*log(lambda)-lambda*sum(X[,2])
plot(lambda,lnL,type="l",xlab=expression(lambda),ylab="log-Liklelihood function")
points(lambdahat,100*log(lambdahat)-lambdahat*sum(X[,2]),pch=17)
```

Run the code below to see how the estimated exponential distribution (full black line) describes the data.

```r
#histogram of the data and estimated exponential distribution
salaries<-seq(0,max(X[,2]),1)
denstity<-lambdahat*exp(-lambdahat*salaries)
hist(X[,2],breaks=100,main="Distribution of salaries",freq=FALSE,xlab="Salaries (CHF)")
lines(salaries,denstity,lwd=2)
```
